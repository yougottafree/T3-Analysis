Hello, in this notebook, I will extract the data from the csv file provided for the state of Arizona only.


In order to do that, we have to read the data from the csv file first. We can do that by first open the file.

```python
data_file = open('us-counties.csv')
```

The file is now saved in the data_file variable, next we will loop through the file and make it into the list of list.

```python
data_list = []
for line in data_file:
    data_list.append(line.strip().split(','))
data_list
```


Next we will loop through the list and write the data in the state of Arizona in the new csv file. 

```python
new_file = open('us-counties-Arizona.csv', 'w')
new_file.write(','.join(data_list[0])+'\n') # the title
for data_row in data_list[1:]:
    if data_row[2] == 'Arizona':
        new_file.write(','.join(data_row)+'\n')
new_file.close()
```

Now the new file is created which contains the information from the state of arizona only. Now we are able to head over to Tableau and visualize the data.
